# Auxiliar function: ISR weights

```python
import pandas as pd
import numpy as np


def merge_event_tables(df_total, df_tt, df_data):
    """
    Fusiona tres tablas 2D (total, tt, data) en una sola por columnas comunes distintas de 'events'.

    Args:
        df_total: DataFrame con columnas ['<bin1>', '<bin2>', 'events']
        df_tt:    DataFrame con columnas ['<bin1>', '<bin2>', 'events']
        df_data:  DataFrame con columnas ['<bin1>', '<bin2>', 'events']

    Returns:
        DataFrame combinado con columnas:
            ['<bin1>', '<bin2>', 'Total', 'MainBgr', 'Data']
    """
    # Detectar automÃ¡ticamente las columnas de binning (todas menos 'events')
    bin_cols = [col for col in df_total.columns if col != "events"]

    # Renombrar columnas de eventos
    df_total_ren = df_total.rename(columns={"events": "Total"})
    df_tt_ren = df_tt.rename(columns={"events": "MainBgr"})
    df_data_ren = df_data.rename(columns={"events": "Data"})

    # Merge sucesivos usando las columnas detectadas
    merged = pd.merge(df_total_ren, df_tt_ren, on=bin_cols, how="outer")
    merged = pd.merge(merged, df_data_ren, on=bin_cols, how="outer")

    # Ordenar por binning si es posible
    merged.sort_values(by=bin_cols, inplace=True)
    merged.reset_index(drop=True, inplace=True)

    return merged
```
