# 2D plot

```python
Y_variable = "ptll"
Y_binning = [0, 20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300, 350, 400, 450, 500]

X_variable = "njets"
X_binning = [0,1, 2, 3, 4, 5] 

df_2D_total = plotter.get_2D_plot_report(X_distribution = X_variable, X_binning = X_binning, 
                           Y_distribution = Y_variable, Y_binning = Y_binning, 
                           background = "total",  
                           include_overflow = overflow, include_underflow = underflow,
                           bin_values = True,
                          )

df_2D_data = plotter.get_2D_plot_report(X_distribution = X_variable, X_binning = X_binning, 
                           Y_distribution = Y_variable, Y_binning = Y_binning, 
                           background = "data", 
                           include_overflow = overflow, include_underflow = underflow,
                            bin_values = True,                                        
                          )

df_2D_tt = plotter.get_2D_plot_report(X_distribution = X_variable, X_binning = X_binning, 
                           Y_distribution = Y_variable, Y_binning = Y_binning, 
                           background = "dy", 
                           include_overflow = overflow, include_underflow = underflow,
                            bin_values = True,                                      
                          )
```

```python
weights_range = (0.9, 1.1) # 0.85-1.10
filter_parameters = (0, 2, 1) # Min; Max, sigma
smoothed = False
df_2D_weights = plotter.get_2D_weights(df_total_bgr = df_2D_total, df_main_bgr = df_2D_tt, df_data = df_2D_data, range = weights_range, filter_parameters = filter_parameters, smoothed = smoothed)
df_2D_weights
```
