# Efficiency

## Efficiency formulas:

$ \varepsilon = \frac{num}{den} $

$\sigma(\varepsilon) = \varepsilon \times \left[ \left(\frac{\sigma_{num}}{num}\right)^2 + \left(\frac{\sigma_{den}}{den}\right)^2  \right]$

## Fit: 

* Root: https://root.cern.ch/root/html524/TMath.html#TMath:Freq


* Python: https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.erf.html#scipy.special.erf


What is the equivalence between both expressions? 

$\int_{-\infty}^{x} e^{-t^2} dt = \int_{-\infty}^{0} e^{-t^2} dt + \int_{0}^{x} e^{-t^2} dt =  \frac{1}{2}  \int_{-\infty}^{\infty} e^{-t^2} dt + \int_{0}^{x} e^{-t^2} dt =  \frac{\sqrt{\pi}}{2} + \frac{\sqrt{\pi}}{2} erf(x) = \frac{\sqrt{\pi}}{2}\left[1 + erf(x)\right]$


Profesor's formula:

$ f(x) = P_{0} +  \frac{P_{1}}{\sqrt{2\pi}} \int_{-\infty}^{\frac{\sqrt{x}- \sqrt{P_{2}}}{2P_{3}}} e^{-\frac{t^2}{2}} = P_{0} + \frac{P_{1}}{\sqrt{2\pi}}\frac{\sqrt{\pi}}{2}\left[1 + erf\left(\frac{\sqrt{x}- \sqrt{P_{2}}}{2P_{3}}\right)\right] = a + b \left[1 + erf\left(\frac{\sqrt{x}- c}{d}\right)\right]   $

Parameters: {a,b,c,d}


- a: Constant.
- b: Amplitude.
- c: mean value $<x>$.
- d: standard deviation $\sigma$.

##### Calculating its derivatives: https://www.youtube.com/watch?v=Jmn8gx3eziI.

* Remember:
  $\frac{d}{dx} \int_{0}^{g(x)} f(t) dt = \frac{d}{dx} [F(g(x)) - F(0)] = \frac{dF(g(x))}{dx} =  f(g(x))\times\frac{dg(x)}{dx}$

$\frac{\partial f}{\partial a} = 1 $

$\frac{\partial f}{\partial b} = \frac{f - a}{b} $ 

$\frac{\partial f}{\partial c} = -\frac{2b}{d\sqrt{\pi}}e^{-\left(\frac{\sqrt{x}-c}{d}\right)^2}$

$\frac{\partial f}{\partial d} =  -\frac{2b(\sqrt{x} - c)}{d^2\sqrt{\pi}} e^{-\left(\frac{\sqrt{x}-c}{d}\right)^2} $




## Variance 

$f(x) =  f(a,b,c,d)$

$\sigma_{f} = \sqrt{ \left(\frac{\partial f}{\partial a}\sigma_{a}\right)^2 + \left(\frac{\partial f}{\partial b}\sigma_{b}\right)^2 +  \left(\frac{\partial f}{\partial c}\sigma_{c}\right)^2 +  \left(\frac{\partial f}{\partial d}\sigma_{d}\right)^2 + 2\left[\frac{\partial f}{\partial a} \frac{\partial f}{\partial b}Cov(a,b) + \frac{\partial f}{\partial a} \frac{\partial f}{\partial c}Cov(a,c) + \frac{\partial f}{\partial a}\frac{\partial f}{\partial d}Cov(a,d)  +  \frac{\partial f}{\partial b} \frac{\partial f}{\partial c}Cov(b,c) +  \frac{\partial f}{\partial b} \frac{\partial f}{\partial d}Cov(b,d) +  \frac{\partial f}{\partial c} \frac{\partial f}{\partial d}Cov(c,d) \right]}$


We can use the matrix form of the covariance matrix:

$ \Sigma = \begin{pmatrix}\sigma _a^2&Cov\left(a,b\right)&Cov\left(a,c\right)&Cov\left(a,d\right)\\ Cov\left(a,b\right)&\sigma _b^2&Cov\left(b,c\right)&Cov\left(b,d\right)\\ Cov\left(a,c\right)&Cov\left(b,c\right)&\sigma _c^2&Cov\left(c,d\right)\\ Cov\left(a,d\right)&Cov\left(b,d\right)&Cov\left(c,d\right)&\sigma _c^2\end{pmatrix} $

$ df = \begin{pmatrix}\frac{\partial f}{\partial a}\\ \frac{\partial \:f}{\partial b}\\ \frac{\partial f}{\partial c}\\ \frac{\partial f}{\partial d}\end{pmatrix} $

Con lo anterior:

$\sigma_{f}^{2} = df^{T} \times \Sigma \times df $
