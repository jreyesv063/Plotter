# QCD shape

```python
import matplotlib.pyplot as plt
import mplhep as hep
import pandas as pd
import numpy as np

bins = np.array(binning)

qcd_map = {
    "CR_B":  plotter.qcd_estimation("cr_b"),
    "CR_C":  plotter.qcd_estimation("cr_c"),
    "CR_D":  plotter.qcd_estimation("cr_d")
}

# Aplicar estilo CMS
plt.style.use(hep.style.CMS)

# Definir colores manualmente (mÃ¡s suaves pero visibles)
colors = {
    "CR_B": "#1f77b4",  # azul
    "CR_C": "#2ca02c",  # verde
    "CR_D": "#d62728",  # rojo
}

# Crear figura y eje
fig, ax = plt.subplots(figsize=(8, 6))

# Dibujar lÃ­neas con grosor consistente y buen espaciado
ax.step(bins, np.append(qcd_map['CR_B'], qcd_map['CR_B'][-1]), where="post", color=colors["CR_B"],
        linewidth=2, label=r"QCD CR$_B$")
ax.step(bins, np.append(qcd_map['CR_C'], qcd_map['CR_C'][-1]), where="post", color=colors["CR_C"],
        linewidth=2, label=r"QCD CR$_C$")
ax.step(bins, np.append(qcd_map['CR_D'], qcd_map['CR_D'][-1]), where="post", color=colors["CR_D"],
        linewidth=2, label=r"QCD CR$_D$")

# Ejes y leyendas
ax.set_xlabel(r"$m_T(\tau, p_T^{miss})$ [GeV]", fontsize=14)
ax.set_ylabel("Events / GeV", fontsize=14)
ax.set_yscale("log")
ax.set_xlim(bins[0], bins[-1])
ax.set_ylim(1, 20000)

# Etiquetas CMS
hep.cms.text("Preliminary", loc=0, fontsize=16)
hep.cms.lumitext("41.5 fb$^{-1}$ (2017, 13 TeV)", fontsize=12)

# Leyenda con tamaÃ±o y separaciÃ³n adecuados
#ax.legend(loc="upper right", fontsize=12, frameon=False)
ax.legend(
    loc="upper center",       # parte superior centrada
    bbox_to_anchor=(0.5, 0.97),  # un poco arriba del Ã¡rea del plot
    ncol=3,                   # nÃºmero de columnas (ajÃºstalo al nÃºmero de entradas)
    fontsize=12,
    frameon=False             # sin borde
)


# Ajustar mÃ¡rgenes
plt.tight_layout()

plt.savefig("qcd_estimation.pdf", format="pdf", bbox_inches="tight")
plt.show()


# ------------------------------
# Crear tabla con valores por bin
# ------------------------------

# Etiquetas de bin (ejemplo: "0.0â€“20.0")
bin_labels = [f"{bins[i]:.1f}â€“{bins[i+1]:.1f}" for i in range(len(bins) - 1)]

# Crear DataFrame con los datos por bin
df_qcd = pd.DataFrame({
    "Bin": bin_labels,
    "CR_B":  plotter.qcd_estimation("cr_b"),
    "CR_C":  plotter.qcd_estimation("cr_c"),
    "CR_D":  plotter.qcd_estimation("cr_d"),
})

# Agregar fila con totales
total_row = pd.DataFrame({
    "Bin": ["Total"],
    "CR_B": [np.sum(qcd_map['CR_B'])],
    "CR_C": [np.sum(qcd_map['CR_C'])],
    "CR_D": [np.sum(qcd_map['CR_D'])],
})

# Concatenar y mostrar tabla
df_qcd = pd.concat([df_qcd, total_row], ignore_index=True)

# Mostrar tabla final
df_qcd
```
