## Stadistical error

### Binomial statistics:

$\delta \varepsilon_{i} = \sqrt{\frac{\epsilon_{i}}{N_{tot}}(1 - \varepsilon_{i}})$

$ N_{i} = L\times \sigma_{i} \times \epsilon_{i} \rightarrow \delta N_{i} = L \times \sigma_{i} \times \delta \varepsilon_{i}$


### Bayesian: 

In order to consider cases where the efficiency is close to 0 or 1 it is neccesary to use Bayesian statistics.


#### Combination

Finally, as the total number of expected events for each background is the combination of different types; for instance, TTToSemiLeptonic + TTToHadronic + TTTo2L2Nu = $tt$.

$ N_{X} = \sum_{i} N_{X_{i}} \rightarrow \delta N_{X} = \sqrt{\sum_{i} \left(\delta N_{X_{i}}\right)^2} $
